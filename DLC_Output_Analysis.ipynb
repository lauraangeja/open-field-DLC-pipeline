### En 
In our case, 50cm corresponds to 455 pixels in the image, so:
- ratio = 0.11 = 0.11cm/1px

In other words, if we have 1 cm, if we want to convert it to pixels, we have to divide by the ratio -> 1cm/0.11=9.1 pixels

If you have 1 pixel, to convert it to centimeters, you have to multiply by the ratio -> 1px*0.11 = 0.11 cm
